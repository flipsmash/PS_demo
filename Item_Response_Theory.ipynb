# Detailed Question Analysis: Item Response Theory

```r
library(ggplot2)
options(repr.plot.width=20, repr.plot.height=12)
library(mirt)
library(ggmirt)
#Read csv file with question cores as correct/wrong (1/0) integers
qscores <- read.csv("C:/Users/brian_local/PS_demo/qscores.csv", colClasses = c("integer"))
```

```r
unimodel <- 'F1 = 1-13'
fit2PL <- mirt(data = qscores, 
               model = unimodel,  # Explain what that means here eventually
               itemtype = "2PL", 
               verbose = FALSE)
```

```r
summary(fit2PL)
```

```r
params2PL <- coef(fit2PL, IRTpars = TRUE, simplify = TRUE)
round(params2PL$items, 2)
```

```r
M2(fit2PL)
```

```r
itemfit(fit2PL)
```

```r
itemfit(fit2PL, fit_stats = "infit")
```

```r
tracePlot(fit2PL)
```

```r
tracePlot(fit2PL,items = c(1:3), facet = F, legend = T) + scale_color_brewer(palette = "Set3")
```

```r
itemInfoPlot(fit2PL) + scale_color_brewer(palette = "Set3")
```
